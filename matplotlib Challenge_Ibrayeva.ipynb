```python
#Creating a figure with two axes
fig2, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

#Setting titles for axises
ax1.set_title('Age Histogram')
ax2.set_title('Political Identity')
#creating histograms for the age and political identity distibution (im not sure whether the pid3 stands for the political identity but thats what google told me )
ax1.hist(x= data['age'], color='black')
ax2.hist(x = data['pid3'], color='purple')
plt.show()
```

```python
#Ceating the bar chart for the evaluation column
# I chose to use the groupby method based on the ideo5 column
evaluation_freq = data.groupby('Evaluation')['ideo5'].count()
evaluation_freq.index
                               
fig3, ax3 = plt.subplots(figsize = (10, 5))
ax3.bar(x= evaluation_freq.index, height = evaluation_freq, color='skyblue')
ax3.set_title('Evaluation of Ideological Position (ideo5)')
plt.show()
```

```python
#creating a scatter plot 
fig4,ax4 = plt.subplots()
ax4.scatter(x=data['ConMediaUse'], y=data['ImmiPhobia'], color='green')
#adding labels and a title 
ax4.set_xlabel('ConMediaUse')
ax4.set_ylabel('ImmiPhobia')
ax4.set_title('Scatter Plot of ConMediaUse vs. ImmiPhobia')
plt.show()
```
