```python
#importing the required packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#choosing the style from the seaborn package
sns.set(style="whitegrid") 
```

Seaborn has five built-in themes to style its plots: darkgrid, whitegrid, dark, white, and ticks. Seaborn defaults to using the darkgrid theme for its plots, but we can change this styling to better suit the presentation needs. 

```python
#Importing the downloaded dataset
Salary_data = pd.read_csv("Salary Data.csv")
```

```python
#Viewing the first 6 rows of data to see its structure
Salary_data.head()
```

```python
#Creating a simple scatterplot with the seaborn default settings 
plt.figure()
sns.scatterplot(x="Age", y="Salary", data=Salary_data)
#adding a title
plt.title('Default Scatter Plot: Age vs Salary')
plt.show()
```

```python
#Creating a scatter plot Age vs Salary and adding additional information for the Education Level and Gender
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data)
plt.title( 'Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
#Defining our palette for the Education Level 
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
# Creating a scatter plot with the customized palette
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors)
plt.title( 'Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show() 
```

# Seaborn markers
* "o": Circle,
* "s": Square,
* "D": Diamond,
* "^": Upward-pointing triangle,
* "v": Downward-pointing triangle,
* "<": Left-pointing triangle,
* ">": Right-pointing triangle,
* "p": Pentagon,
* "*": Star
* "+": Plus sign
* "x": Cross

```python
# Defining custom color palette
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
# Defining custom markers for the Gender
gender_markers = {"Female": "o", "Male": "s"}

# Creating a scatter plot with custom colors and markers
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors, markers=gender_markers)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
#Changing the size of the shape and getting rid of grid lines
# Defining custom color palette
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
# Defining custom markers
gender_markers = {"Female": "o", "Male": "s"}

# Creating a scatter plot with custom colors and markers
plt.figure(figsize=(12, 8)) #figures size
sns.set_style("ticks") #no gridlines
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors, markers=gender_markers)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
#Customazing the legend
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}

plt.figure(figsize=(12, 8))
sns.set_style("ticks") 
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors, markers=gender_markers)
#removing the legend frame
plt.legend(frameon=False)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
# Dividing the legend in 2 rows and changing its location
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}
plt.figure(figsize=(12, 8))
sns.set_style("ticks") 
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors, markers=gender_markers)

#removing the legend frame and dividing it in 2 columns
plt.legend(ncol=2, frameon=False, loc= "lower right" )
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
# Creating a separate scatter plot for each Education Level for better clarity in visualization - BACHELORS
#Since we won't have a hue for Education level, we will define a custom pallete for the gender now
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks") #no gridlines
Bachelors=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Bachelor's"], palette=gender_colors)
# Setting limits for x and y axes
Bachelors.set_xlim(25, 50)  
Bachelors.set_ylim(50000, 250000)
#customazing the legend
plt.legend(frameon=False, loc= "upper left" )
plt.title("Age vs Salary Based on Gender Among Bachelor's Degree Holders")
plt.show()
```

```python
# Creating a separate scatter plot for each Education Level for better clarity in visualization -MASTERS
#Since we won't have a hue for Education level, we will define a custom pallete for the gender now
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks") #no gridlines
Masters=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Master's"], palette=gender_colors)

# Setting limits for x and y axes
Masters.set_xlim(25, 50)  
Masters.set_ylim(50000, 250000)
#customazing the legend
plt.legend(frameon=False, loc= "upper left" )
plt.title("Age vs Salary Based on Gender Among Master's Degree Holders")
plt.show()
```

```python
# Creating a separate scatter plot for each Education Level for better clarity in visualization - PHD
#Since we won't have a hue for Education level, we will define a custom pallete for the gender now
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks") #no gridlines
phd=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "PhD"], palette=gender_colors)
# Setting limits for x and y axes
phd.set_xlim(25, 50)  
phd.set_ylim(50000, 250000)
#customazing the legend
plt.legend(frameon=False, loc= "upper left" )
plt.title("Age vs Salary Based on Gender Among PhD Degree Holders")
plt.show()
```

```python
## Creating a FacetGrid with subplots for each education level
#defining the palette for the gender
gender_colors = {"Female": "red", "Male": "blue"}

g = sns.FacetGrid(Salary_data, col="Education Level")
g.map_dataframe(sns.scatterplot, x= "Age", y= "Salary", hue="Gender", palette=gender_colors)  
#Adjusting the legend
g.add_legend(title="Gender", frameon=False, loc="lower right")

plt.show()
```

```python
#Creating a figure with 3 axis
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
gender_colors = {"Female": "red", "Male": "blue"}
sns.set_style("ticks") #no gridlines
# Use Seaborn's scatterplot for the fourth subplot
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Bachelor's"], palette=gender_colors, ax=ax[0])
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Master's"], palette=gender_colors, ax=ax[1])
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "PhD"], palette=gender_colors, ax=ax[2])

for axes in ax:
    axes.legend(frameon=False)

# Set the title in the center
plt.suptitle('Scatter Plot: Age vs Salary by Education Level and Gender', ha="center")

# Show the plot
plt.show()
```

```python
#Creating a figure with 3 axis and setting the limits
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
gender_colors = {"Female": "red", "Male": "blue"}
sns.set_style("ticks") #no gridlines
scatter_1 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Bachelor's"],
                            palette=gender_colors, ax=ax[0])
scatter_2 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Master's"],
                            palette=gender_colors, ax=ax[1])
scatter_3 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "PhD"],
                            palette=gender_colors, ax=ax[2])

plt.suptitle('Scatter Plot: Age vs Salary by Education Level and Gender', ha="center")
# Setting the limits for x and y axes
for axes in ax:
    axes.set_xlim(25, 50)  
    axes.set_ylim(50000, 250000) 
    axes.legend(frameon=False, loc = "upper left")

# Show the plot
plt.show()
```

## Creating bar charts in seaborn

```python
#Creating a bar chart 
sns.barplot(x='Years of Experience', y='Education Level', hue='Gender',data=Salary_data)
```

```python
# Adding cap on error bar
sns.barplot(x='Education Level', y='Years of Experience', hue='Gender', capsize=0.1, data=Salary_data)
```

```python
#Removing the error bar 
sns.barplot(x='Education Level', y='Years of Experience', hue='Gender', capsize=0.1, data=Salary_data, errorbar=None)
```

```python
#changing the hue order and colors
sns.barplot(x='Education Level', y='Years of Experience', hue='Gender', capsize=0.1, data=Salary_data, errorbar=None, palette='dark:blue', hue_order=['Female','Male'])
```
